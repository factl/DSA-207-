```python
pivot1 = df.pivot_table(index=['Gender'], values=['Age', 'proficiency in English and Japanese', 'satisfied with WiFi and Internet','satisfied with library'], aggfunc=np.mean)
Qcut1 = pd.qcut(pivot1['proficiency in English and Japanese'], q=4)
bin_labels_4 = ['bad','normal', 'good','very good' ]
df['quartile_EJ'] = pd.qcut(df['proficiency in English and Japanese'],q=[0, 0.05, 0.25, 0.75, 1],labels=bin_labels_4)


#countplot
ax = sns.countplot(data=df, x='quartile_EJ')
ax.tick_params(axis='x', labelrotation=90)
```
